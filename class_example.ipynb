## Class Example

```python
import matplotlib.pyplot as plt
class iocurve ():
    def __init__(self, fig, ax, x, y):
        self.x = x
        self.y = y
        self.colors = 'blue'
        self.ax = ax
        self.ax.scatter (x, y, color = self.colors)
        self.x_title = 'Intensity (mA)'
        self.y_title = 'Slope (mV)'
        
           
```

```python
intensity = [0, 50, 100, 150, 200, 250, 300, 350]
slope = [0.00881, 0.27161, 0.11727, 0.15019, 0.21035, 0.27902, 0.3418, 0.35618]

fig, ax = plt.subplots(figsize=(4,4), dpi = 100)

exp = iocurve (fig, ax, intensity, slope)



```

```python
fig, ax = plt.subplots(figsize=(4,4), dpi = 100)
exp2 = iocurve (fig, ax, intensity, slope)
exp2.ax.spines['top'].set_visible(False)
exp2.ax.spines['right'].set_visible(False)
exp2.ax.set_xlabel(exp.x_title, size = 12)
exp2.ax.set_ylabel(exp.y_title, size = 12)


```
